# Importing libraries

```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

# Importing the dataset

```python
#Importing the dataset 
credit = pd.read_csv('ExploratoryAnalysis.csv', header = 0)
#Removing extra ID column
credit = credit[credit.columns[1:len(credit.columns)]]
#Visualizing 
credit.describe()
```

```python
#Including the y variable at the end of the dataframe
DEFAULT = credit['DEFAULT']
del credit['DEFAULT']
credit = pd.concat([credit, DEFAULT], axis=1)
credit.head()
```

# Preprocessing

```python
#Rearranging the order of the columns
colnames = credit.columns.tolist()
colnames =  ['ID', 'LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'CurBill_April','CurBill_May','CurBill_June',
             'CurBill_July','CurBill_Aug', 'CurBill_Sept', 'PayStat_April', 'PayStat_May', 'PayStat_June', 'PayStat_July',
             'PayStat_Aug', 'PayStat_Sept', 'DEFAULT']
credit = credit[colnames]
```

## Has the customer defaulted before?

```python
credit['Risky'] = (np.where((credit['PayStat_Sept'] >= 1) |
                           (credit['PayStat_Aug'] >= 1) |
                           (credit['PayStat_July'] >= 1) |
                           (credit['PayStat_June'] >= 1) |
                           (credit['PayStat_May'] >= 1) |
                           (credit['PayStat_April'] >= 1), 'Yes', 'No'))
```

```python
#What is the ratio of customers that have defaulted before and are predicted as defaulting in October?
print("Of the people who are predicted to default in October,",
      (len(credit[((credit.DEFAULT == "defaulted") & (credit.Risky == "Yes"))]) / 
      len(credit[credit.DEFAULT == "defaulted"]))*100,
      "% has defaulted at least once in the previous months as well.")
```

# Calculating the credit ratio

```python
#Create variable that provides ratio of current bill and amount that can be borrowed
credit['CREDIT_RATIO'] = (credit['CurBill_Aug'] / credit['LIMIT_BAL'])

#Check for the difference in the ratio people that default and people that pay
defaulters = credit[credit.DEFAULT == "defaulted"]
nondefaulters = credit[credit.DEFAULT == "paid"]

print("Average credit ratio is", credit.CREDIT_RATIO.mean())
print("Average credit ratio for defaulters is", defaulters.CREDIT_RATIO.mean())
print("Average credit ratio for non-defaulters is", nondefaulters.CREDIT_RATIO.mean())
```

## Averages of the bills and amounts paid

```python
#Generate a variable that gives the value for the average amount of the bill over the months
credit['MeanBill'] = ((credit.CurBill_April +
            credit.CurBill_May +
            credit.CurBill_June +
            credit.CurBill_July +
            credit.CurBill_Aug +
            credit.CurBill_Sept) / 6)

#Generate a variable that gives the value for the average amount paid over the months
credit['MeanPaid'] = ((credit.PayStat_April +
            credit.PayStat_May +
            credit.PayStat_June +
            credit.PayStat_July +
            credit.PayStat_Aug +
            credit.PayStat_Sept) / 6)
```

```python
credit.head()
```

# Selecting the variables to keep

```python
#Keeping the variables that seem important for predicting whether the customer
#will default next month
credit = credit[['LIMIT_BAL', 'MeanBill','MeanPaid','CurBill_Sept','Risky','CREDIT_RATIO','AGE','EDUCATION','DEFAULT']]
```

```python
#Remove the observations where the individual was categorized as defaulted even if he/she had no debt in September
credit = credit[-((credit.CurBill_Sept == 0) & (credit.DEFAULT == 'defaulted'))]
```
